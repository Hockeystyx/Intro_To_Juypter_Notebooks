# What is Jupyter Notebook?

Jupyter Notebook is an open-source web application that allows you to create and share documents that contain live code, equations, visualizations and narrative text. Uses include: data cleaning and transformation, numerical simulation, statistical modeling, data visualization, machine learning, and much more.

## Opening Notebook

When you first open Jupyter Notebook you will see a screen similar to below. Open(running) notebooks will appear green, while inactive notebooks are gray. Clicking on the name will open the notebook in a new tab.

## New Notebook

To create a new notebook click the New button and select Python 3.

Your new notebook will open in a new tab.  Click on Untitled and rename your notebook.

## What is a Kernel?

A notebook kernel is a computational engine that executes the code contained in a Notebook document. We are using the ipython kernel and running Python 3. When you open a Notebook document, the associated kernel is automatically launched.

Running a notebook is usually a top down process.  One can import all the modules needed in the first code cell and use them in any cell that follows.  Same with asigning variables or functions.  When you restart the kernel you reset all variables and lose access to all modules or variables above the cell selected and will need to rerun those cells if referenced in the cells below. When trying to diagnos a problem(bug), the numbers In [1]: show the order the cells are executed in. 

### Options for running the Notebook:

1. Step-by-step (one cell at a time) by pressing the run button or control + enter.
2. Whole notebook in a single step by clicking on the menu Cell -> Run All.
3. Restart the kernel and run whole notebook by click on the menu Kernel -> Restart & Run All. 

### Reseting the Notebook:

Under the menu Kernel -> Restart & Clear Output will reset your notebook deleting all output.


## Properly closing a Notebook.

Notebooks will remain open until shutdown. Closing a tab doesn't close the notebook! There are a few ways to do this and best practices is to only have one notebook running at a time. The easiest is by clicking on the menu File -> Close and Halt.

Alternitively under the Running tab of the main dashboard you can click the Shutdown button.
